```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/Iris.csv')
```

```python
data.sample(10)
```

```python

#Conversion of list into array
SL=np.array(data['SepalLengthCm'])
SW=np.array(data['SepalWidthCm'])
PL=np.array(data['PetalLengthCm'])
PW=np.array(data['PetalWidthCm'])
```

```python
#Shape of all features
print(f"Shape of Sepal Length is {SL.shape}")
print(f"Shape of Sepal Width is {SW.shape}")
print(f"Shape of Petal Length is {PL.shape}")
print(f"Shape of Petal Width is {PW.shape}")
```

```python
#Sorting of all features
print("Sorted Sepal Length:")
print(np.sort(SL))
print("")
print("Sorted Sepal Width:")
print(np.sort(SW))
print("")
print("Sorted Patal Length:")
print(np.sort(PL))
print("")
print("Sorted Patal Width:")
print(np.sort(PW))
print("")
```

```python
#Finding data type
print(f"Data type of Petal Length is {PL.dtype}")
print(f"Data type of Petal Width is {PW.dtype}")
print(f"Data type of Sepal Length is {SL.dtype}")
print(f"Data type of Sepal Width is {SW.dtype}")
```
